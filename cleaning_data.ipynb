```python
# Add the Pandas dependency
import pandas as pd
import os
```

```python
# File to load
file_to_load = os.path.join ("Resources", "missing_grades.csv")

# Reading csv file into DataFrame
missing_grade_df = pd.read_csv(file_to_load)
missing_grade_df
```

```python
# Calculating the average reading score for males: 
# code taken from https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
df = pd.DataFrame(data = missing_grade_df)
df.groupby("gender", as_index=False).reading_score.mean()
```

```python
# Calculating the average math score for males:
# code taken from https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
df.groupby("gender", as_index=False).math_score.mean()
```

```python
# Fill in the empty reading score with average reading score for all males
# code taken from https://stackoverflow.com/questions/13842088/set-value-for-particular-cell-in-
# pandas-dataframe-using-index
df.at[2,"reading_score"] = 77.1
```

```python
# Fill in the empty math score with average math score for all males
# code taken from https://stackoverflow.com/questions/13842088/set-value-for-particular-cell-in-
# pandas-dataframe-using-index
df.at[5,"math_score"] = 64.5
df
```
